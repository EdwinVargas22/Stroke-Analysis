```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as st
```

```python
# Import Stroke Dataset 
stroke_df = pd.read_csv("Clean_Resources/stroke_dataset.csv")

stroke_df.head()
```

```python
# Set the catplot
sns.set_style(style="darkgrid")
sns.set_context("poster")

g = sns.catplot(
    data=stroke_df, kind="bar",
    x="hypertension", y="stroke", hue="gender",
    ci=None, palette="dark", alpha=.8, height=8
)

# Settting the Labels
(g.set_axis_labels("", "Avg count of Stroke Column")
    .set_xticklabels(["No Hypertension", "Yes Hypertension"]))

g.fig.suptitle("Avg Stroke Count per Hypertension", y=1.03)
plt.xticks(rotation=90)
plt.show()

# Saving the bar chart
g.savefig('Resources/Avg_Stroke_Hypertension.png')
```

```python
# Set the boxplot
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_context("notebook")

sns.boxplot(x="stroke", y="age",
            hue="gender", palette=["m", "g"],
            data=stroke_df)

sns.despine(offset=10, trim=True)
plt.title('Correlation between Age and Stroke', fontsize=30)
plt.legend(loc='best', fontsize=20)
plt.xticks([0, 1], ['No Stroke', 'Stroke'], fontsize=20)
plt.xlabel('')
plt.ylabel('Age', fontsize=20)

# Saving the boxplot
plt.savefig('Resources/Correlation_Age_Stroke.png')
plt.show()
```

```python
# Set the boxplot
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_context("notebook")

sns.boxplot(x="stroke", y="avg_glucose_level",
            hue="gender", palette=["m", "g"],
            data=stroke_df)

sns.despine(offset=10, trim=True)

plt.title('Correlation between Avg Glucose Levels and Stroke', fontsize=28)
plt.legend(loc='best', fontsize=20)
plt.xticks([0, 1], ['No Stroke', 'Stroke'], fontsize=20)
plt.xlabel('')
plt.ylabel('Avg Glucose Levels', fontsize=20)

# Saving the boxplot
plt.savefig('Resources/Correlation_Glucoselevels_Stroke.png')
plt.show()
```

```python
# Set the boxplot
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_context("notebook")

sns.boxplot(x="stroke", y="bmi",
            hue="gender", palette=["m", "g"],
            data=stroke_df)

sns.despine(offset=10, trim=True)
plt.title('Correlation between Body Mass Index and Stroke', fontsize=28)
plt.legend(loc='best', fontsize=20)
plt.xticks([0, 1], ['No Stroke', 'Stroke'], fontsize=20)
plt.xlabel('')
plt.ylabel('Body Mass Index', fontsize=20)

# Saving the boxplot
plt.savefig('Resources/Correlation_BMI_Stroke.png')
plt.show()
```

```python
# Set the heatmap
plt.figure(figsize=(10,10))
sns.set_context("paper", rc={"font.size":50,"axes.titlesize":20,"axes.labelsize":20})

# Setting the tick names for the x and y
xtick_names = ['Age', 'Hypertension', 'Heart Disease', 'Avg Glucose Level', 'Body Mass Index', 'Stroke']
ytick_names = ['Age', 'Hypertension', 'Heart Disease', 'Avg Glucose Level', 'Body Mass Index', 'Stroke']

g = sns.heatmap(stroke_df.corr(), 
            vmin=-1,
            cmap='coolwarm',
            annot=True, 
            annot_kws={'size':15},
            xticklabels= xtick_names,
            yticklabels= ytick_names);

plt.title("Correlations")
cbar = g.collections[0].colorbar
cbar.ax.tick_params(labelsize=15)
plt.xticks(rotation=75, fontsize="15")
plt.yticks(np.arange(6)+0.5, rotation=0, fontsize="15")

# Saving the heatmap
plt.savefig('Resources/Correlations.png')
plt.show()
```
