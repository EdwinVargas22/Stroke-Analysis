```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as st
```

```python
# Import Stroke Dataset 
stroke_df = pd.read_csv("Clean_Resources/stroke_dataset.csv")

stroke_df.head()
```

```python
# sns.set_theme(style="whitegrid")

# Draw a nested barplot by stroke and gender
sns.set_style(style="whitegrid")

g = sns.catplot(
    data=stroke_df, kind="bar",
    x="stroke", y="avg_glucose_level", hue="gender",
    ci=None, palette="dark", alpha=.8, height=8
)

# Settting the Labels
(g.set_axis_labels("", "Avg Glucose Levels")
    .set_xticklabels(["No Stroke", "Stroke"]))
```

```python
sns.set_style(style="whitegrid")

g = sns.catplot(
    data=stroke_df, kind="bar",
    x="stroke", y="hypertension", hue="gender",
    ci=None, palette="dark", alpha=.8, height=8
)

# Settting the Labels
(g.set_axis_labels("", "Hypertension")
    .set_xticklabels(["No Stroke", "Stroke"]))
plt.show()
```

```python
sns.set_style(style="whitegrid")

g = sns.catplot(
    data=stroke_df, kind="bar",
    x="stroke", y="heart_disease", hue="gender",
    ci=None, palette="dark", alpha=.8, height=8
)

# Settting the Labels
(g.set_axis_labels("", "Heart Disease")
    .set_xticklabels(["No Stroke", "Stroke"]))
plt.show()
```
